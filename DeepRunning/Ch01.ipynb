```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def sigmoid_function(x):
    return 1/(1+np.exp(-x))
x = np.linspace(-5, 5)
y = sigmoid_function(x)
plt.plot(x, y)
plt.show()
```

```python
def tanh_function(x):
    return np.tanh(x)
x=np.linspace(-5,5)
y=tanh_function(x)

plt.plot(x,y)
plt.show()
```

```python
def relu_function(x):
    return np.where(x<=0,0,x)
x=np.linspace(-5,5)
y=relu_function(x)

plt.plot(x,y)
plt.show()
```

```python
def leaky_rrelu_function(x):
    return np.where(x<=0,0.01*x,x)
x=np.linspace(-5,5)
y=leaky_rrelu_function(x)

plt.plot(x,y)
plt.show()
```

```python
def softmax_function(x):
    return np.exp(x)/np.sum(np.exp(x)) # 
y = softmax_function(np.array([1,2,3]))
print(y)
```

```python
#  

#x1,x2 

X1=np.arange(-1.0,1.0,0.2)
X2=np.arange(-1.0,1.0,0.2)

#  10*10
Z=np.zeros((10,10))

#x1,x2   
W_x1=2.5
W_x2=3.0

#
bias=0.1

#     
for i in range(10):
    for j in range(10):
        
#    +         
        u=X1[i]*W_x1+X2[j]*W_x2+bias
    
#   
        y=1/(1+np.exp(-u)) # 
        Z[j][i]=y
        
#  
plt.imshow(Z,"gray",vmin= 0.0, vmax=1.0)
plt.colorbar()
plt.show()
```

```python
# x, y 
X = np.arange(-1.0, 1.0, 0.1) #   20
Y = np.arange(-1.0, 1.0, 0.1)

# 
w_im = np.array([[1.0,2.0],
[2.0,3.0]]) #  2x2 
w_mo = np.array([[-1.0,1.0],
[1.0,-1.0]]) #  2x2 

# 
b_im = np.array([0.3,-0.3]) # 
b_mo = np.array([0.4,0.1]) # 

# 
def middle_layer(x, w, b):
    u = np.dot(x, w) + b
    return 1/(1+np.exp(-u)) #  

# 
def output_layer(x, w, b):
    u = np.dot(x, w) + b
    return np.exp(u)/np.sum(np.exp(u)) #  

#    
x_1 = []
y_1 = []
x_2 = []
y_2 = []

#     
for i in range(20):
    for j in range(20):
        
# 
        inp = np.array([X[i], Y[j]])
        mid = middle_layer(inp, w_im, b_im)
        out = output_layer(mid, w_mo, b_mo)
        
#    
        if out[0] > out[1]:
            x_1.append(X[i])
            y_1.append(Y[j])
        else:
            x_2.append(X[i])
            y_2.append(Y[j])
            
#  
plt.scatter(x_1, y_1, marker="+")
plt.scatter(x_2, y_2, marker="o")
plt.show()
```
